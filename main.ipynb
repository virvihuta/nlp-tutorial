### Bag Of Words (BOW)

`"I love the book"`

`"This is a great book"`

`"The book is great"`

`"I love the shoes"`

Using the BOW we extract every unique word as such:

`"I love the book this is a great fit shoes"`

Then we create a *Matrix* that represents wether each document has the word.

```python
train_x = ["i love the book", "this is a great book", "the book is great", "i love the shoes"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(train_x)

print(vectors.toarray())
```

This is the *Matrix* that represents wether the document has the word.
